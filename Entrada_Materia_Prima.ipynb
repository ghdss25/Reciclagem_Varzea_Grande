## Importação das Bibliotecas 

```python
import pandas as pd
import numpy as np 
import matplotlib as mat 
import matplotlib.pyplot as plt 
import seaborn as sns
import colorsys
plt.style.use('seaborn-talk') 
import warnings 
warnings.filterwarnings("ignore") 
%matplotlib inline

import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as py
```

## Leitura dos Dados 

```python
lancamentos = pd.read_excel("Entrada_Materia_Prima.xlsx", sheet_name="Lançamentos MP")

lancamentos
```

```python
lancamentos = lancamentos.loc[:, ~lancamentos.columns.str.contains('^Unnamed')]

lancamentos
```

```python
lancamentos = lancamentos.dropna(subset=['Supervisor', 'Placa', 'LOTE', 'NF-e', 'Fornecedor', 'Produto', 
                                         'Situacao', 'HORA CHEGADA', 'Peso Recebido', 'Peso NF-e'])
lancamentos
```

```python
## Verificação de dados nulos
lancamentos.isna().sum()
```

```python
## Verificação de dados duplicados
lancamentos.duplicated().sum()
```

```python
## Lista de todas as colunas 
list(lancamentos)
```

```python
## 1 - Distribuição de Lotes
lancamentos.LOTE.hist(bins = 616)
plt.xlabel("LOTES") 
plt.ylabel("Número de Lotes") 
plt.title("Distribuição de Lotes")
plt.show()
```

```python
# 2 - Qual a distribuição de produtos, esta mais duro ou mole ? 

## definindo a quantidade 
labels = lancamentos.Situacao.value_counts().index
num = len(lancamentos.Produto.value_counts().index) 

## Criando lista de cores
listaHSV = [(x*1.0/num, 0.5, 0.5) for x in range(num)] 
listaRGB = list(map(lambda x: colorsys.hsv_to_rgb(*x), listaHSV))

# Gráfico da Pizza 
fatias, text = plt.pie(lancamentos.Situacao.value_counts(), colors = listaRGB, startangle = 90) 
plt.legend(fatias, labels, bbox_to_anchor = (1.05,1)) 
plt.title("Situação")
plt.show()
```

```python
## 3 - Crie um contador que mostre o número de fornecedores pertencentes a cada Peso Recebido 
fornecedor_peso = lancamentos.groupby(["Fornecedor"]).count()["Peso Recebido"].rename("Total")

analise_fornecedor = pd.DataFrame({"Total" : fornecedor_peso})

analise_fornecedor
```

```python
## 4 - Crie um plot que mostre o número de produtos pertencentes 
sns.set_style("whitegrid")
g = sns.catplot(y="Produto", data=lancamentos, kind="count", palette="tab10", height=7, aspect=1.5) 
g.ax.set_title("Produtos", fontdict={'size': 18})
g.ax.xaxis.set_label_text("Número de Produtos", fontdict = {'size':16}) 
g.ax.yaxis.set_label_text("Produto", fontdict={'size':16})
plt.show()
```

```python
## 5 - Crie um plot com o Peso do Produto na visualização de supervisor. 
fig, ax = plt.subplots(figsize=(8,5))
colors = ["#6A5ACD", "#191970", "#708090"]
sns.barplot(x="Produto", y="Peso Recebido", hue="Supervisor", palette=colors, data=lancamentos) 
ax.set_title("Peso Recebido dos Produtos visualizados por Supervisores", fontdict={'size':12})
ax.xaxis.set_label_text("Produtos", fontdict= {'size':12})
ax.yaxis.set_label_text("Peso Recebido", fontdict = {'size':12})
plt.show()
```

```python
## 6 - Calcule a média de Peso NF-e por Produto e Situação 

trial = pd.DataFrame() 
for p in list(lancamentos["Produto"].unique()):
    
    for s in list(lancamentos["Situacao"].unique()):
        
        t = lancamentos[(lancamentos["Produto"] == p) & (lancamentos["Situacao"] == s)]["Peso NF-e"].mean()
        trial = trial.append(pd.DataFrame({'Produto':p, 'Situacao':s, 'Peso NF Médio':t}, index=[0])) 

trial = trial.reset_index() 
del trial["index"] 
trial["Peso NF Médio"].fillna(0, inplace=True) 
trial["Peso NF Médio"].isnull().value_counts()
trial["Peso NF Médio"] = trial["Peso NF Médio"].astype(int) 
trial.head(5)
```

```python
## Criando um heatmap com a média de Peso NF-e por Produto e Situação  
tri = trial.pivot("Produto", "Situacao", "Peso NF Médio") 
fig, ax = plt.subplots(figsize=(15,20))
sns.heatmap(tri, linewidths=1, cmap="summer", annot=True, ax=ax, fmt="d")
ax.set_title("Heatmap - Peso NF Médio com Situação em Produto", fontdict={'size':20})
ax.xaxis.set_label_text("Tipo de Situação", fontdict={'size':20}) 
ax.yaxis.set_label_text("Produto", fontdict={'size':20})
plt.show()
```

## Leitura dos Dados do Oleo_Saturado 

```python
oleo_saturado = pd.read_excel("Entrada_Materia_Prima.xlsx", sheet_name="Oleo Saturado")

oleo_saturado
```

```python
oleo_saturado = oleo_saturado.dropna(subset=['Supervisor'])
```

```python
oleo_saturado
```

## Modificação para Mês, Dia e Ano 

```python
oleo_saturado["Dia"] = pd.DatetimeIndex(oleo_saturado["Data"]).day
oleo_saturado["Ano"] = pd.DatetimeIndex(oleo_saturado["Data"]).year
oleo_saturado["Mês"] = pd.DatetimeIndex(oleo_saturado["Data"]).month

oleo_saturado
```

```python
## Deletando a coluna Data
oleo_saturado = oleo_saturado.drop(columns=['Data'])
```

```python
oleo_saturado
```

```python
## 1 - Crie um Count Plot que mostre o número de Peso recebidos a cada supervisor 

fig = px.bar(
    
    oleo_saturado,
    x ='Dia', 
    y ='Peso Recebido', 
    title ='Peso recebidos a cada supervisor', 
    color ='Supervisor')

fig.show()
```

```python
## 2 - Agrupamento de Peso Recebido por Dia em Lote
pesos_dia = oleo_saturado.groupby(['Peso Recebido', 'Dia']).mean()['LOTE'].reset_index()
```

```python
## 2 - Gráfico : Quantidade de Peso Recebido por dias em Lotes
plt.figure(figsize=(30,8)) 
sns.lineplot(x="Dia", y="LOTE", data=pesos_dia)
```

```python
## 3 - Comportamento dos Produtos pelo Recebimento dos Pesos recebidos em relação aos dias de entregue
```

```python
plt.figure(figsize=(20, 10)) 
oleo_saturado['Peso Recebido'] = oleo_saturado['Peso Recebido'] + oleo_saturado['Peso Recebido'] 

sns.lineplot(x="Dia", y="Peso Recebido", hue='Produto', data=oleo_saturado)
```

```python
## 4 - Informe a media de Peso por Mês e Ano 
```

```python
ano_mes_peso_medio = oleo_saturado.groupby(["Mês", "Ano"]).mean()["Peso Recebido"].rename('Média de Peso') 

analise_geral_peso = pd.DataFrame({"Média de Peso" : ano_mes_peso_medio})

analise_geral_peso["Média de Peso"] = analise_geral_peso["Média de Peso"].map("{:.2f}".format)

analise_geral_peso
```

```python
## 5 - Quantidade e Total de Peso Recebido, por Placas de Carros das Reciclagens
oleo = go.Bar(x = oleo_saturado['Placa'], 
              y = oleo_saturado['Peso Recebido']) 

data = [oleo] 
py.plot(data)
```

## Leitura dos dados da Planilha Sangue Liquido 



```python
sangue_liquido = pd.read_excel("Entrada_Materia_Prima.xlsx", sheet_name="Sangue Liquido")

sangue_liquido
```

## Modificação para Dia, Mês, Ano 

```python
sangue_liquido["Dia"] = pd.DatetimeIndex(sangue_liquido["Data"]).day
sangue_liquido["Ano"] = pd.DatetimeIndex(sangue_liquido["Data"]).year
sangue_liquido["Mês"] = pd.DatetimeIndex(sangue_liquido["Data"]).month

sangue_liquido
```

```python
sangue_liquido
```

```python
## 1 - Crie um gráfico que mostre os superiores que obtiveram mais pesos recebidos dos produtos 

pesos_produtos_superiores = pd.DataFrame(sangue_liquido.Supervisor.value_counts().rename_axis('Supervisor').reset_index(
    
    name='Peso Recebido').head(10)) 

pesos_produtos_superiores
```

```python
plt.figure(figsize=(10,7)) 
sns.barplot(x='Supervisor', y = 'Peso Recebido', data = pesos_produtos_superiores)
plt.ylabel('Peso Recebido', size = 13) 
plt.xlabel('Supervisor', size = 13)
plt.title('Supervisor com maior número de Pesos', size = 15) 
plt.yticks(fontsize=12) 
plt.xticks(fontsize=12, rotation = 80) 
plt.show()
```

```python
## 2 - Organize a frequência dos Pesos Realizados por Dia 
sangue_liquido['Dia'].value_counts()
```

```python
plt.figure(figsize=(10,7))
sns.histplot(data = sangue_liquido, x = 'Dia', bins=35, kde=True) 
plt.ylabel('Pesos Realizados', size = 13) 
plt.xlabel('Dia', size = 13) 
plt.title('Frequência anual de Pesos Realizados', size = 15) 
plt.yticks(fontsize=12) 
plt.xticks(fontsize=12, rotation=80) 
plt.show()
```

```python
## 3 - Crie um gráfico que mostre o maior número de pesos recebidos, pelos supervisores
sangue_liquido.groupby('Supervisor')[['Peso Recebido']].sum().sort_values('Peso Recebido')
```

```python
## Visualizando os pesos recebidos dos supervisores 
plt.figure(figsize=(10,7))
sns.lineplot(data = sangue_liquido.groupby('Supervisor')[['Peso Recebido']].sum(), 
            x = 'Supervisor', y = 'Peso Recebido', color = 'green', marker = 'o') 

plt.ylabel('Pesos Recebidos por Supervisores', size = 13) 
plt.xlabel('Supervisor', size = 13) 
plt.title('Número de Pesos Realizados por Supervisores', size = 15) 
plt.yticks(fontsize = 12) 
plt.xticks(fontsize = 12, rotation=80) 
plt.show()
```

```python
## 4 - Quantidade de Lotes, Placas por Pesos Realizados 
lotes_placas_pesos = sangue_liquido.groupby(["LOTE", "Placa"]).sum()["Peso Recebido"].rename("Total de Peso")

analise_geral_ltp = pd.DataFrame({"Total de Peso" : lotes_placas_pesos})

analise_geral_ltp.head(32)
```

```python
display(analise_geral_ltp.sort_values("Total de Peso", ascending = False))
```

```python
f, ax = plt.subplots(figsize=(20,20))
sns.lineplot(x='LOTE', y='Total de Peso', hue='Placa', data=analise_geral_ltp)
plt.xticks(rotation=360)
```

```python
# 5 - Informe a média de Peso Realizado do Ano de 2022, com o Mês e Dia

m = sangue_liquido.groupby(["Dia", "Mês", "Ano"]).mean()["Peso Recebido"].rename("Média")

analise_a_p_d_m = pd.DataFrame({"Média" : m})

analise_a_p_d_m["Média"] = analise_a_p_d_m["Média"].map("{:.2f}".format)

analise_a_p_d_m
```

```python
f, ax = plt.subplots(figsize=(20,20))
sns.lineplot(x='Dia', y='Média', hue='Mês', data=analise_a_p_d_m)
plt.xticks(rotation=360)
```
