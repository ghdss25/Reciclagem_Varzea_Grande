## Importação de Bibliotecas

```python
import pandas as pd 

import datetime
```

## Leitura dos dados de Depósito 

```python
deposito = pd.read_excel("Controle_Entrada_Materia_Prima.xlsx", sheet_name="DEPOSITO")

deposito
```

## Removendo dados Nulos com o nome Unnamed

```python
deposito = deposito.loc[:, ~deposito.columns.str.contains('^Unnamed')]

deposito
```

## Modificação para Mês, Dia e Ano 

```python
deposito["Dia"] = pd.DatetimeIndex(deposito["Data"]).day
deposito["Ano"] = pd.DatetimeIndex(deposito["Data"]).year
deposito["Mês"] = pd.DatetimeIndex(deposito["Data"]).month

deposito
```

```python
deposito = deposito.drop(columns=['Data'])
```

```python
deposito
```

## Verificação de Dados Duplicados 

```python
deposito.duplicated().sum()
```

## Análise de Peso por Fornecedor

```python
peso_fornecedor = deposito.groupby(["Fornecedor"]).sum()["Peso Recebido"].rename("Total de Peso")
media_peso_fornecedor = deposito.groupby(["Fornecedor"]).mean()["Peso Recebido"].rename("Média de Peso")

analise_peso_fornecedor = pd.DataFrame({"Total de Peso" : peso_fornecedor, 
                                        "Média de Peso" : media_peso_fornecedor})

analise_peso_fornecedor["Média de Peso"] = analise_peso_fornecedor["Média de Peso"].map("{:.2f}".format)

analise_peso_fornecedor
```

```python
display(analise_peso_fornecedor.sort_values("Total de Peso", ascending=False).head(24)[ \
    ['Total de Peso', 'Média de Peso']])
```

## Análise de Peso por Produto 

```python
peso_produto = deposito.groupby(["Produto"]).sum()["Peso Recebido"].rename("Total de Peso")
media_peso_produto = deposito.groupby(["Produto"]).mean()["Peso Recebido"].rename("Média de Peso")

analise_peso_produto = pd.DataFrame({"Total de Peso" : peso_produto, 
                                        "Média de Peso" : media_peso_produto})

analise_peso_produto["Média de Peso"] = analise_peso_produto["Média de Peso"].map("{:.2f}".format)

analise_peso_produto
```

## Peso total por Ano e Mês

```python
Ano_Mes_Peso = deposito.groupby(["Ano", "Mês", "Dia"]).sum()["Peso Recebido"].rename("Total de Peso") 
Ano_Mes_Peso_Media = deposito.groupby(["Ano", "Mês", "Dia"]).mean()["Peso Recebido"].rename("Média de Peso") 

analise_ano_mes = pd.DataFrame({"Total de Peso" : Ano_Mes_Peso, 
                                "Média de Peso" : Ano_Mes_Peso_Media})

analise_ano_mes["Média de Peso"] = analise_ano_mes["Média de Peso"].map("{:.2f}".format)

analise_ano_mes
```

## Lote, Placa por Peso 

```python
lote_placa_peso = deposito.groupby(["Lote", "Placa"]).sum()["Peso Recebido"].rename("Total de Peso") 
lote_placa_medio_peso = deposito.groupby(["Lote", "Placa"]).mean()["Peso Recebido"].rename("Média de Peso")

analise_lote_placa_peso = pd.DataFrame({"Total de Peso" : lote_placa_peso, 
                                        "Média de Peso" : lote_placa_medio_peso})

analise_lote_placa_peso["Média de Peso"] = analise_lote_placa_peso["Média de Peso"].map("{:.2f}".format)

analise_lote_placa_peso.head(60)
```

## Planilha Expedição PA 

```python
expedicao_pa = pd.read_excel("Controle_Entrada_Materia_Prima.xlsx", sheet_name="EXPEDICAO_PA")

expedicao_pa
```

## Removendo dados Nulos com o nome Unnamed

```python
## Apagar a Tabela Data
expedicao_pa = expedicao_pa.loc[:, ~expedicao_pa.columns.str.contains('^Unnamed')]

expedicao_pa
```

```python
expedicao_pa["Dia"] = pd.DatetimeIndex(expedicao_pa["Data"]).day
expedicao_pa["Ano"] = pd.DatetimeIndex(expedicao_pa["Data"]).year
expedicao_pa["Mês"] = pd.DatetimeIndex(expedicao_pa["Data"]).month

expedicao_pa
```

```python
## Apagar a Tabela Data
expedicao_pa = expedicao_pa.drop(columns=['Data'])

```

```python
expedicao_pa
```

## Análise da Expedição PA de Placa e Lote por Peso 

```python
placa_lote_peso = expedicao_pa.groupby(["Placa", "LOTE"]).sum()["PESO"].rename("Total") 

analise_lote_peso_geral = pd.DataFrame({"Total" : placa_lote_peso})

analise_lote_peso_geral.head(10)
```

## Análise da expedição PA CLIENTE / DESTINO por Peso

```python
cliente_destino_motorista_peso = expedicao_pa.groupby(["CLIENTE_DESTINO", "MOTORISTA"]).sum()["PESO"].rename("Total de Peso")  

analise_geral = pd.DataFrame({"Total de Peso" : cliente_destino_motorista_peso})

analise_geral
```

## Análise de carregamento e produto por peso total

```python
carregamento_produto = expedicao_pa.groupby(["Carregamento", "Produto"]).sum()["PESO"].rename("Total de Peso") 

analise_geral = pd.DataFrame({"Total de Peso" : carregamento_produto})

analise_geral
```
